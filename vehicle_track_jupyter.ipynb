```python
-
import sqlite3
import os
from flask import Flask, jsonify
from flask_cors import CORS


app = Flask(__name__)
CORS(app)
BASE_DIR = os.path.abspath(os.getcwd())
DB_PATH = os.path.join(BASE_DIR, 'lojistik_v2.db') # Yeni veritabanı dosyası
PORT = 5055

def db_kurulum():
    if os.path.exists(DB_PATH):
        os.remove(DB_PATH)
    
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    
    
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS arac_loglari (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            plaka TEXT, enlem REAL, boylam REAL, hiz INTEGER, zaman TEXT
        )
    ''')
    
    veriler = []
    
   
    veriler.extend([
        ('34 MBZ 3939', 41.0168, 28.9750, 20, '09:00'), # Eminönü
        ('34 MBZ 3939', 41.0201, 28.9782, 45, '09:02'), # Galata Köprüsü
        ('34 MBZ 3939', 41.0235, 28.9815, 50, '09:04'), # Karaköy
        ('34 MBZ 3939', 41.0260, 28.9850, 55, '09:06'), # Tophane
        ('34 MBZ 3939', 41.0295, 28.9895, 40, '09:08'), # Kabataş
        ('34 MBZ 3939', 41.0330, 28.9930, 30, '09:10'), # Dolmabahçe
        ('34 MBZ 3939', 41.0370, 28.9980, 60, '09:12'), # Stad
        ('34 MBZ 3939', 41.0405, 29.0025, 50, '09:14'), # Beşiktaş
        ('34 MBZ 3939', 41.0440, 29.0080, 0,  '09:16')  # İskele
    ])
    
    
    veriler.extend([
        ('35 MB 3939', 38.4061, 27.1086, 30, '09:00'), # Üçyol Metro
        ('35 MB 3939', 38.4120, 27.1250, 60, '09:03'), # Yeşildere Yolu Başı
        ('35 MB 3939', 38.4185, 27.1450, 70, '09:06'), # Atatürk Maskı Hizas
        ('35 MB 3939', 38.4250, 27.1650, 80, '09:09'), # Kemer Altgeçit
        ('35 MB 3939', 38.4350, 27.1850, 75, '09:12'), # Otogar Sapağı
        ('35 MB 3939', 38.4450, 27.2050, 50, '09:15'), # Bornova Girişi
        ('35 MB 3939', 38.4520, 27.2150, 40, '09:17'), # Küçükpark Yakını
        ('35 MB 3939', 38.4580, 27.2270, 20, '09:19'), # Ege Üni Kampüs Girişi
        ('35 MB 3939', 38.4620, 27.2300, 0,  '09:21')  # Fen Fakültesi Önü
    ])
    
    cursor.executemany('INSERT INTO arac_loglari (plaka, enlem, boylam, hiz, zaman) VALUES (?,?,?,?,?)', veriler)
    conn.commit()
    conn.close()
    

@app.route('/api/tum-araclar', methods=['GET'])
def veri_getir():
    try:
        conn = sqlite3.connect(DB_PATH)
        conn.row_factory = sqlite3.Row
        cursor = conn.cursor()
        
        
        cursor.execute("SELECT * FROM arac_loglari ORDER BY zaman ASC")
        rows = cursor.fetchall()
        conn.close()
        
        sonuc = [dict(row) for row in rows]
        return jsonify(sonuc), 200
    except Exception as e:
        return jsonify({"error": str(e)}), 500

if __name__ == '__main__':
    db_kurulum()
    print(f"Sunucu {PORT} portunda aktif...")
    app.run(port=PORT, debug=True, use_reloader=False)
```
